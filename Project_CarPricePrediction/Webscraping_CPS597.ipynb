## Web Scraping of enterprisecarsales.com

```python
from bs4 import BeautifulSoup
import csv
import json
import re
import requests

# Create a list to store the extracted data
data_list = []

car_count = 0

# Iterate over the desired range of pages
for page in range(1, 101):
    url = f"https://www.enterprisecarsales.com/list/buy-a-car/srp-page-{page}/?zipcode=48858"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    script_tags = soup.find_all('script', id='rocket-script-b4ddd57-grid')

    for script_tag in script_tags:
        script_content = script_tag.string
        json_objects = re.findall(r"window.jzlSetVehicleInfoContext\('(\w+)', (\{.*?\})\);", script_content)

        for vehicle_id, json_str in json_objects:
            car_count += 1
            data = json.loads(json_str)

            Name = data['make']
            Model = data['make'] + " "+ data['model']
            Year = data['year']
            Stock_number = data['stockNumber']
            Mileage = data['mileage']
            Price = data['price']
            Fuel_Type = data['fuelType']
            MPG_City = data['mpgCity']
            MPG_Highway = data['mpgHighway']
            Vehicle_Type = data['vehicleType']
            
            # Append the extracted data to the list
            data_list.append([car_count, vehicle_id, Name, Model, Year, Stock_number, Mileage, Fuel_Type, MPG_City, MPG_Highway, Vehicle_Type, Price])

            print(f"Car Count: {car_count}")
            print(f"Vehicle ID: {vehicle_id}")
            print(f"Name: {Name}")
            print(f"Model: {Model}")
            print(f"Year: {Year}")
            print(f"Stock Number: {Stock_number}")
            print(f"Mileage: {Mileage}")
            print(f"Fuel Type: {Fuel_Type}")
            print(f"MPG City: {MPG_City}")
            print(f"MPG Highway: {MPG_Highway}")
            print(f"Vehicle Type: {Vehicle_Type}")
            print(f"Price: {Price}")
            print()

# Define the path and filename for the CSV file
csv_file = 'carpricepred_output.csv'

# Write the data to the CSV file
with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    # Write the header row
    writer.writerow(['Car Count', 'Vehicle ID', 'Name', 'Model', 'Year', 'Stock Number', 'Mileage', 'Fuel Type', 'MPG City', 'MPG Highway', 'Vehicle Type', 'Price'])
    # Write the data rows
    writer.writerows(data_list)

print(f"Data saved to {csv_file}.")
```
